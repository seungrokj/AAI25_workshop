# Step 0 Run useful functions for vLLM SERVER

```python
def launch_vllm(envs, args):
    import multiprocessing, time
    import subprocess

    vllm_run = f'''
       VLLM_LOGGING_LEVEL=ERROR \
       {envs} \
       vllm serve RedHatAI/Llama-3.1-8B-Instruct \
            --disable-log-requests \
            --trust-remote-code -tp 1 \
            {args} \
            --cuda-graph-sizes 64 \
            --chat-template /app/vllm/examples/tool_chat_template_llama3.1_json.jinja \
        '''
    def vllm_call():
        subprocess.run(vllm_run, shell=True, check=True)
    multiprocessing.Process(target=vllm_call).start()
```

# STEP 1 vLLM v0 Server launch

From vLLM server, you should see these messages first. The execute the followings

<span style="color:red"> *To stop the server just press "interrupt the kernel"*</span>

```python
envs = "VLLM_USE_V1=0"
args = ""
launch_vllm(envs, args)
```

# STEP 2 vLLM v1 Server launch

From vLLM server, you should see these messages first. The execute the followings

<span style="color:red"> *To stop the server just press "interrupt the kernel"*</span>

```python
envs = "VLLM_USE_V1=1 VLLM_V1_USE_PREFILL_DECODE_ATTENTION=1 "
args = "--no-enable-prefix-caching"
launch_vllm(envs, args)
```

# STEP 3 vLLM v1 + Prefix Caching Server launch

From vLLM server, you should see these messages first. The execute the followings

<span style="color:red"> *To stop the server just press "interrupt the kernel"*</span>

```python
envs = "VLLM_USE_V1=1 VLLM_V1_USE_PREFILL_DECODE_ATTENTION=1 "
args = "--enable-prefix-caching"
launch_vllm(envs, args)
```
